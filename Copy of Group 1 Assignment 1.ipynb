import random: Imports the random module for generating random numbers.

```python
import random
```

Dice Rolling:
roll_dice(): Function that simulates rolling two six-sided dice and returns a tuple of two integers representing the dice rolls.

```python
def roll_dice():
    return random.randint(1, 6), random.randint(1, 6)
```

Calculating Sum of Dice:

calculate_sum(dice): Function that takes a tuple dice (representing two dice rolls) and returns the sum of the two numbers.

```python
def calculate_sum(dice):
    return sum(dice)
```

Checking Win/Loss/Point:

->check_win(dice_sum): Function that checks the outcome based on the sum of the dice:

->If the sum is 7 or 11, returns "Win".

->If the sum is 2, 3, or 12, returns "Lose".

->Otherwise, returns "Point".

```python
def check_win(dice_sum):
    if dice_sum in [7, 11]:
        return "Win"
    elif dice_sum in [2, 3, 12]:
        return "Lose"
    else:
        return "Point"
```

Handling Making a Point:

make_point(point): Function that handles the phase where a point has been established:
Continues rolling the dice until either the point is rolled again (returns "Win") or a 7 is rolled (returns "Lose").

```python
def make_point(point):
    while True:
        new_dice_sum = calculate_sum(roll_dice())
        if new_dice_sum == point:
            return "Win"
        elif new_dice_sum == 7:
            return "Lose"
```

Playing the Game:

play_game(): Function that orchestrates a complete round of Craps:

* Prints a welcome message

* Rolls the dice.

* Calculates the sum of the dice.

* Determines the initial outcome (Win, Lose, or Point).

If a Point is established, continues to handle subsequent rolls until a final result is determined.

```python
def play_game():
    print("Welcome to Craps!")
    print("You roll the dice and get: ")
    dice = roll_dice() # Call the roll_dice() function to get the dice values
    print(dice)

    dice_sum = calculate_sum(dice)
    print(f"The sum of the dice is: {dice_sum}")
    result = check_win(dice_sum)
    print(f"{result}")

    if result == "Point":
        point = dice_sum
        print(f"Your point is: {point}")
        result = make_point(point)
        print(f"And your final result is: {result}")
```

Execution:

The play_game() function is called at the end to initiate and play a single round of Craps.

```python
play_game()
```
